# HDFS Commands Examples

**Command**: mkdir

**Example**: bin/hdfs dfs -mkdir /user/ec2-user/input



**Command**: ls

**Example**: bin/hdfs dfs -ls /user/ec2-user/



**Command**: rmdir

**Example**: bin/hdfs dfs -rmdir /user/ec2-user/input



**Command**: put

**Example**:  bin/hdfs dfs -put sftp://ec2-user@ec2-54-67-106-32.us-west-1
.compute.amazonaws.com/home/ec2-user/dsci551/hello.txt /user/ec2-user/input



**Command**: get

**Example**:  bin/hdfs dfs -get /user/ec2-user/input/hello.txt sftp://ec2-
user@ec2-54-67-106-32.us-west-1.compute.amazonaws.com/home/ec2-user/dsci551/hw1





**Command**: cp

**Example**: bin/hdfs dfs -cp /user/ec2-user/input /user/ec2-user/output



**Command**: cat

**Example**: bin/hdfs dfs -cat /user/ec2-user/input/hello.txt



**Command**: rm

**Example**: bin/hdfs dfs -rm /user/ec2-user/input/hello.txt
